```python
def t_to_DeltaE_singel(t_offset, time_data, detector, L2):
    
    detector = detector*1.602176634e-22 # [J]
    
    time = time_data[time_data[:,0]>=40*1e3]
    
    t = time[:,0]*1e-6 # [s]
    m_n = 1.67492749804e-27 # [kg]
    L1 = (162-6.35) # [m]
    L2 = L2#1.1+1.189 # [m]
    
    Ei = (L1**2*m_n)/(2*(t-np.sqrt((m_n*L2**2)/(2*detector))-t_offset)**2)
    Delta_E = (Ei-detector)*6.24150907e21 # convert from J to meV
    Ef = Delta_E-Ei
    
    mu = np.average(Delta_E, weights=time[:,1])
    FWHM = np.sqrt(np.average((Delta_E-mu)**2, weights=time[:,1]))*2.3548
    
    #print('µ = ', mu)
    #print('var = ', FWHM)
    
    return np.column_stack((Delta_E, time[:,1::])), mu, FWHM

L2_2p7 = 1.1 + 1.189
L2_3p2 = 1.238 + 1.316
L2_3p8 = 1.342 + 1.42
L2_4p4 = 1.443 + 1.521
L2_5p0 = 1.557 + 1.623
```

### Fitting gauss for a detector

```python
def gauss(x,a,mu,sigma):
    return  a*np.exp(-(x-mu)** 2 / (2*sigma**2))

delta_E = reduced_2p7_0mev[0]
I = reduced_2p7_0mev[1]
I_err = reduced_2p7_0mev[2]

chi2fit_delta_E = Chi2Regression(gauss, delta_E[I>0.005], I[I>0.005], I_err[I>0.005]) #Chi2Regression(gauss, qh_in13[tot_cnt>1], tot_cnt[tot_cnt>1], tot_cnt_err[tot_cnt>1]) # Chi**2 fit to the histograms excluding the empty bins.
minuit_delta_E = Minuit(chi2fit_delta_E, a=0.035, mu=4, sigma=0.005)
minuit_delta_E.limits["a"] = [0,0.05]
minuit_delta_E.limits["sigma"] = [0.003,1]

minuit_delta_E.migrad();
fit_parameters_delta_E = minuit_delta_E.values
fit_errors_delta_E = minuit_delta_E.errors
Chi2_delta_E = minuit_delta_E.fval

print('FWHM=',2.35*fit_parameters_delta_E[2])

print(minuit_delta_E.hesse())

print('chi=',Chi2_delta_E)

x_range = np.linspace(np.min(delta_E), np.max(delta_E),1000000)

plt.title('4 meV on 2.7 meV analyzer')
plt.errorbar(reduced_2p7_0mev[0], reduced_2p7_0mev[1], yerr=reduced_2p7_0mev[2], label='2.7 meV detector \n FWHM = 0.02 meV', color='#00a1df')
plt.plot(x_range, gauss(x_range, *fit_parameters_delta_E), label='Gaussian fit', color='#D37E30', lw=3)
plt.xlim(3.925,4.075)
plt.xlabel('$\Delta E$ [meV]')
plt.ylabel('I [a.u.]')
plt.grid()
plt.legend(fontsize=10)
plt.savefig('4mev_gauss_fit_0p1ms.png')
```
